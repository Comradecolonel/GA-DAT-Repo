<h1>Model Tuning</h1>

```python
rf.get_params()
```

```python
samples_per_leaf = [1, 5, 10, 15, 25, 50]
rf_scores = []


for num_samples in samples_per_leaf:
    rf.set_params(min_samples_leaf=num_samples)
    scores5 = cross_val_score(estimator=rf, X=X_train, y=y_train, cv=10)
    rf_scores.append((np.mean(scores5), num_samples))
```

```python
rf_scores
```

```python
rf.set_params(min_samples_leaf=2)
```

```python
n_estimators = [10, 25, 50, 100, 200]
rf_scores = []


for numestimators in n_estimators:
    rf.set_params(n_estimators=numestimators)
    scores5 = cross_val_score(estimator=rf, X=X_train, y=y_train, cv=10)
    rf_scores.append((np.mean(scores5), numestimators))
```

```python
rf_scores
```

```python
rf.set_params(n_estimators = 100)
```

```python
features = [0.2, 0.4, 0.5, 0.6, 0.7]
rf_scores = []


for n_features in features:
    rf.set_params(max_features=n_features)
    scores5 = cross_val_score(estimator=rf, X=X_train, y=y_train, cv=10)
    rf_scores.append((np.mean(scores5), n_features))
```
