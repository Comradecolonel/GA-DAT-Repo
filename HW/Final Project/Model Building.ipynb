```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

df = pd.read_csv('productsprocessed1.csv')
#standardize the data
#df=(df-df.mean())/df.std()
#standardizing actually gave me a lower score on the final test (around .18)
```

```python
df
```

```python
df_train = df.iloc[1000:,:]
df_test = df.iloc[:1000,:]
```

```python
#importing 
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge, Lasso

rf = RandomForestRegressor()
rreg= Ridge()
lasso = Lasso()
```

```python
X_train = df_train.drop('price',axis = 1)
y_train = df_train['price']
```

```python
from sklearn.model_selection import cross_val_score
#you can use fit and predict method from other models beside sklearn
scores = cross_val_score(estimator=rf, X=X_train, y=y_train, cv=10)
```

```python
scores.var()
```

```python
scores.mean()
```

```python
scores2 = cross_val_score(estimator=rreg, X=X_train, y=y_train, cv=10)
```

```python
scores2.var()
```

```python
scores2.mean()
```

```python
scores3 = cross_val_score(estimator=lasso, X=X_train, y=y_train, cv=10)
```

```python
scores3.var()
```

```python
scores3.mean()
```

```python
#Ridge regression did the best, let's see if we can improve it.

rreg.get_params()
```

```python
alphas = np.logspace(-4, 4, 9)
cv_scores = []

for alpha in alphas:
    rreg.set_params(alpha=alpha)
    scores = cross_val_score(estimator=rreg, X=X_train, y=y_train, cv=10)
    cv_scores.append((np.mean(scores), alpha))
```

```python
cv_scores
```

```python
rreg.set_params(alpha=1000)
```

```python
scores4 = cross_val_score(estimator=rreg, X=X_train, y=y_train, cv=10)
```
