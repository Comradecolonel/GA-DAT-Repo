```python
#getting rid of irrelevant columns
data2 = data2.drop(['id', 'reviews_url','tags', 'url', 'discount_price','title','metascore'],  axis = 1)
```

```python
#Need to parse out string to just get the names of the developers
devs = data2['developer']
```

```python
dev_list = []
for items in devs:
    if type(items) is not float and 'Publisher:' in items:
        items = items.split('Publisher:')
        if 'Franchise:' in items[0]:
            items = items[0].split('Franchise:')
            dev_list.append(items[0])
        else:
            dev_list.append(items[0])
    elif type(items) is not float and 'Release Date:' in items:
        items = items.split('Release Date:')
        if 'Franchise:' in items[0]:
            items = items[0].split('Franchise:')
            dev_list.append(items[0])
        else:
            dev_list.append(items[0])
    else:
        items = "Other"
        dev_list.append(items)
```

```python
#need to the do the same with publishers. Not all games have a publisher so I used the developers column since it has the publisher label in it
pubs = data2['developer']
```

```python
pub_list = []
for items in pubs:
    if type(items) is not float and 'Publisher:' in items:
        items = items.split('Publisher:')
        items = items[1].split('Release Date:')
        if 'Franchise:' in items[0]:
            items = items[0].split('Franchise:')
            pub_list.append(items[0])
        else:
            pub_list.append(items[0])
    else:
        items = "None"
        pub_list.append(items)
```

```python
#need to seperate out the dates and convert them to a standardized format
dates = data2['developer']
```

```python
date_list = []
for items in dates:
    if type(items) is not float and 'Release Date:' in items:
        items = items.split('Release Date:')
        date_list.append(items[1])
    else:
        items = 'None'
        date_list.append(items)
```

```python
date_list
```

```python
#There are some dates that aren't formatted right but will still have a month name in it. It's usually a planned release date
date_list_fin = []
abb_dates = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
for date in date_list:
    if  any(substring in date for substring in abb_dates):
        try:
            date = pd.to_datetime(date)
            date_list_fin.append(date)
        except:
            date_list_fin.append(0)
    else:
        date_list_fin.append(0)
```

```python
len(date_list_fin)
```

```python
#Just a simple encoding for whether or not a game is still in development
data2['early_access'] = data2['early_access'].replace(False, 0) 
data2['early_access'] = data2['early_access'].replace(True, 1) 
```

```python
#need to format prices and set all free games to a price of zero. if the data isn't available its set to 0
prices = []
for items in data2['price']:
    if type(items) is str:
        prices.append(0)
    elif items is np.nan:
        prices.append(0)
    else:
        prices.append(items)
```

```python
#Used ordinal encoding as a proxy for a rating number. If there is no rating its set to a neutral 5

sentiment_list = []

for items in data2['sentiment']:
    if items == 'Overwhelmingly Positive':
        sentiment_list.append(9)
    elif items == 'Very Positive':
        sentiment_list.append(8)
    elif items == 'Mostly Positive':
        sentiment_list.append(7)
    elif items == 'Positive':
        sentiment_list.append(6)
    elif items == 'Mixed':
        sentiment_list.append(5)
    elif items == 'Mostly Negative':
        sentiment_list.append(4)
    elif items == 'Negative':
        sentiment_list.append(3)
    elif items == 'Very Negative':
        sentiment_list.append(2)
    elif items == 'Overwhelmingly Negative':
        sentiment_list.append(1)
    else:
        sentiment_list.append(5)
    
```

```python
#this was just a test to see how it would look
data2['sent_transformed'] = sentiment_list
```

```python
data2 = data2.drop('sent_transformed',axis=1)
```

```python
for item in data2['genres']:   
    print(type(item))
```

```python
#Here we generate a list of genres that we one-hot encode using a mult-label method. Each value in the genre column is a list so we can't just use getdummies
genre_list = []

for items in data2['genres']:
    if type(items) is list:
        genre_list.append(items)
    else:
        genre_list.append(['None'])
        
data2['genres'] = genre_list
```

```python
genre_list
```

```python
genres = data2['genres']

mlb = MultiLabelBinarizer()

genres_dummies = pd.DataFrame(mlb.fit_transform(genres),columns=mlb.classes_, index=data2.index)

genres_dummies = genres_dummies.drop('None', axis =1)
```

```python
#We are doing the same thing with specs since it is also a list with limited possible values
specs_list = []

for items in data2['specs']:
    if type(items) is list:
        specs_list.append(items)
    else:
        specs_list.append(['None'])
        
data2['specs'] = specs_list
```

```python
specs_list
```

```python
specs = data2['specs']

spec_dummies = pd.DataFrame(mlb.fit_transform(specs),columns=mlb.classes_, index=data2.index)

spec_dummies = spec_dummies.drop('None', axis =1)
```

```python
data2.columns
```

```python
#Here is where the magic happens. These lines will replace the columns with all the cleaned values

data2['developer'] = dev_list
data2 = data2.join(genres_dummies)
data2['price'] = prices
data2['publisher'] = pub_list
data2['release_date'] = date_list_fin
data2['sentiment'] = sentiment_list
data2 = data2.join(spec_dummies)
```

```python
data2 = data2.drop(['genres','specs'], axis = 1)
```
