```python
diff_days = []
for date in df['release_date']:
    if len(date) != 1:
        date = pd.to_datetime(date)
        t_date = current_date - date
        t_date = str(t_date)
        t_date = t_date.split('days')[0]
        t_date = int(t_date)
        diff_days.append(t_date)
    else:
        date = int(date)
        diff_days.append(date)
```

```python
for date in diff_days:
        print(type(date))
```

```python
print(pd.Series(diff_days).mean())
```

```python
diff_days_fin = []
for date in diff_days:
    if date == 0:
        diff_days_fin.append(972)
    else:
        diff_days_fin.append(date)
```

```python
diff_days_fin
```

```python
df['days_released'] = diff_days_fin
```
